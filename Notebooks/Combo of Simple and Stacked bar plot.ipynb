## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# setting default parameters
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['axes.titlesize'] = 20 
plt.rcParams['legend.fontsize'] = 16  
```

## Loading files

```python
data = pd.read_csv('dataset.csv')
```

```python
data.head()
```

## Ploting

```python
fig, ax = plt.subplots(figsize = (16, 10))

w = 0.3
data[data.columns[data.columns.str.contains('(?:Low-Med)')]].plot(kind='bar', stacked = True, color = ['indianred', 'dodgerblue', 'goldenrod'], ax = ax, width = w)

for bar in ax.patches:
    bar.set_x(bar.get_x() - w)

data[data.columns[data.columns.str.contains('(?:Med-High)')]].plot(kind='bar', stacked = True, color = ['indianred', 'dodgerblue', 'goldenrod'], ax = ax, width = w, alpha = 0.5)

for bar in ax.patches:
    bar.set_x(bar.get_x() + w/2)

ax.set_xticklabels(data["region"], rotation = 90)
ax.set_xlabel("Region")
ax.set_ylabel("Cost (billion USD/yr.)")
ax.set_title('Average Annual Cost for Water Sector (High-Medium Reliability)')

plt.savefig(r'\Images\combo_of_simple_and_stacked_bar.png', dpi = 300, bbox_inches='tight')

plt.show()
```
